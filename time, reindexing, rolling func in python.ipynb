# reindexing

```python
import numpy as np

column=['a','b','c','d','e']
index=['A','B','C','D','E']
df1=pd.DataFrame(np.random.rand(5,5),columns=column, index=index)

print(df1)
print("\n\ DataFrame after reindexing rows: \n")
print(df1.reindex(['B','D','A','C','E']))
```

```python
NEW_INDEX=['U','A','B','C','R']
print(df1.reindex(NEW_INDEX))
```

```python
column=['e','a','b','c','d']
print(df1.reindex(column, axis='columns'))
```

```python
column=['g','a','b','c','r']
print(df1.reindex(column, axis='columns'))
```

```python
column=['g','a','b','c','r']
print(df1.reindex(column, axis='columns', fill_value=1.5))
```

```python
column=['g','a','b','c','r']
print(df1.reindex(column, axis='columns', fill_value='data missing'))
```

# rolling function

```python
import pandas as pd
import numpy as np

dates=pd.date_range(start='2022-01-01', end='2022-01-31', freq='d')
numbers=np.random.randint(100,size=(31))

df=pd.DataFrame({'Dates':dates, 'Price':numbers})
df.head(10)
```

```python
df.plot.line(x='Dates',y='Price')
```

```python
df['Moving Average']=df['Price'].rolling(window=5).mean()
df.head(10)
```

```python
df.plot.line(x='Dates',y='Moving Average')
```

```python
df.plot.line(x='Dates',y=['Price','Moving Average'])
```

```python
df['Rolling Average center']=df['Price'].rolling(window=5, center=True).mean()
df.head(10)
```

```python
df.iloc[5:7,1]=np.NaN
df
```

```python
df['Min Periods'] = df['Price'].rolling(5,min_periods=5).mean()
df
```

#### Handling Timezone in python

```python
import pytz
import datetime

dt_local=datetime.datetime.now()
dt_utc=datetime.datetime.now(pytz.utc)

print(dt_local,'\n',dt_utc)
```

```python
dt_utc=datetime.datetime.now(pytz.timezone('US/Eastern'))
print(dt_utc)
```

```python
dt_asia=datetime.datetime.now(pytz.timezone('Asia/Kolkata'))
print(dt_asia)
```

```python
dt_turkey=datetime.datetime.now(pytz.timezone('Turkey'))
print(dt_turkey)
```

```python
dt_local=datetime.datetime.now()
dt_usc=dt_local.astimezone(pytz.timezone('US/Central'))
print(dt_local,'\n',dt_usc)
```

```python
for timezone in pytz.all_timezones:
    print(timezone)
```
