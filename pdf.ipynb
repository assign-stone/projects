```python
#creating function to get the whole text

def full_text(file_path):
    try:
        doc = fitz.open(file_path)
        ReadPDF = PyPDF2.PdfReader(file_path)
        
        the_whole_text = ""
        
        for i in range(pages):
            one_page = ReadPDF.pages[i] #.pages[page_number]
            text = one_page.extract_text()
            the_whole_text += text
            
            sent = sent_tokenize(the_whole_text)
            
            words = word_tokenize(the_whole_text)
        
        return the_whole_text, sent, words
    
    except FileNotFoundError:
        print("File not found.")
        return None
    except Exception as e:
        print(f"Error occurred {str(e)}")
        return None
        
```

```python
the_whole_text, sent, words = full_text(file_path)
```

```python
#another way or functions to use for same task
def extract_pdf_text(file_path):
    with open(file_path, 'rb') as pdf_file:
        ReadPDF = PyPDF2.PdfReader(pdf_file)
        
        text = ""
        
        for p in range(len(ReadPDF.pages)): #len(ReadPDF.pages)
                page = ReadPDF.pages[p]
                page_text = page.extract_text()
                text += page_text
        return text

text= extract_pdf_text(file_path)

if text:
    print(text)
```

```python
import fitz  # PyMuPDF
from summarizer import Summarizer

# Function to extract text from a PDF
def extract_text_from_pdf(file_path):
    text = ""
    doc = fitz.open(file_path)
    for page_num in range(doc.page_count):
        page = doc.load_page(page_num)
        text += page.get_text()
    doc.close()
    return text

# Function to generate a summary
def generate_summary(text):
    summarizer = Summarizer()
    summary = summarizer(text)
    return summary

```

```python
if __name__ == "__main__":
    file_path = input("upload pdf file: ")  
    extracted_text = extract_text_from_pdf(file_path)
    summary = generate_summary(extracted_text)
    print(summary)
```
